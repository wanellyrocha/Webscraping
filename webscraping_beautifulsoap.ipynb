#### Identificando os pa√≠ses e suas capitais

```python
import requests

html = requests.get('https://www.scrapethissite.com/pages/simple/').text

html_parsed = BeautifulSoup(html, 'html.parser')
```

```python
print(html_parsed.prettify())
```

```python
html_parsed.find(class_ = 'country-name').text.strip()    # usando o 'strip' para retirar os espa√ßos
```

```python
html_parsed.find(class_ = 'country-capital').text.strip()    # usando o 'strip' para retirar os espa√ßos
```

#### Usando um loop para identificar pa√≠s e capital de todos da p√°gina

```python
paises = html_parsed.find_all(class_ = 'country-name')
capitais = html_parsed.find_all(class_ = 'country-capital')

lista = []

for pais, capital in zip(paises, capitais):
    dic = {'pais' : pais.text.strip(),
           'capital' : capital.text.strip()}
    
    lista.append(dic)
```
